```python
from pymongo import MongoClient
from pprint import pprint
import pandas as pd
import matplotlib.pyplot as plt
```

```python
mongo = MongoClient(port=27017)
db = mongo.Project3
```

```python
print(db.list_collection_names())
```

```python
print(db.data1.find_one())
```

```python
# Get the MongoDB collection, replace "collection_name" with your collection name
collection = db["data1"]

# Get all documents from the collection
cursor = collection.find()
data = list(cursor)

# Convert the list of documents to a pandas DataFrame
df = pd.DataFrame(data)

# Convert the column to integer, replace "field_name" with your field name
df["data_value"] = pd.to_numeric(df["data_value"], errors='coerce', downcast='integer')
df["low_confidence_limit"] = pd.to_numeric(df["low_confidence_limit"], errors='coerce', downcast='integer')
df["high_confidence_limit"] = pd.to_numeric(df["high_confidence_limit"], errors='coerce', downcast='integer')
# Drop the "_id" column before loading the data back to MongoDB
df.drop(columns=["_id"], inplace=True)

# Delete the existing data in the MongoDB collection
collection.delete_many({})

# Load the updated data to the MongoDB collection
collection.insert_many(df.to_dict("records"))
```
